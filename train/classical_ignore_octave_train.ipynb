# 分析 MIDI 文件

```python
import os
from data import __PATH__
from data.preprocess import analyse_all_midi

midi_file_path = os.path.join(__PATH__, 'classical')
processed_midi_dataset = [x for x in analyse_all_midi(midi_file_path)]
processed_midi_dataset[:10]
```

# MIDI 向量化编码

```python
from data.note_digitize import encode as d_encode
from data.note_vectorize import encode as v_encode

dataset = []
for _ in processed_midi_dataset:
    dataset.append((_[0], [x[0] for x in _[1] if '_' in x[0] and x[0][-1].isdigit()]))
dataset = [(_[0], [v_encode(d_encode(x)) for x in _[1]]) for _ in dataset]
dataset[0][1][:10]
```

# 数据处理，处理为自回归格式

```python
from data.auto_regression_preprocess import auto_regression_preprocess

n_gram = 5
auto_regression_dataset = [(_[0], auto_regression_preprocess(_[1], n_gram)) for _ in dataset]
auto_regression_dataset[0][1][0], auto_regression_dataset[0][1][1]
```

# 数据集参数设置

```python
import torch

BATCH_SIZE = 2
DEVICE = torch.device('cuda')
DTYPE = torch.float32
DATA_SLICE = 5000
```

# 创建数据集加载器

```python
import torch
from data.auto_regression_preprocess import get_auto_regression_dataset_loaders

auto_regression_dataset_inputs = auto_regression_dataset[0][1][0].tolist()
auto_regression_dataset_labels = auto_regression_dataset[0][1][1].tolist()
for _ in auto_regression_dataset[1:]:
    auto_regression_dataset_inputs.extend(_[1][0].tolist())
    auto_regression_dataset_labels.extend(_[1][1].tolist())

auto_regression_dataset_inputs = torch.tensor(auto_regression_dataset_inputs[:DATA_SLICE], dtype=DTYPE, device=DEVICE)
auto_regression_dataset_labels = torch.tensor(auto_regression_dataset_labels[:DATA_SLICE], dtype=DTYPE, device=DEVICE)
print(auto_regression_dataset_inputs, auto_regression_dataset_labels)
print(auto_regression_dataset_inputs.shape, auto_regression_dataset_labels.shape)
auto_regression_dataset_train_loader, auto_regression_dataset_valid_loader = get_auto_regression_dataset_loaders(auto_regression_dataset_inputs, auto_regression_dataset_labels, train_ratio=0.95, batch_size=BATCH_SIZE)
```

# 初始化自回归模型

```python
from deeplotx import AutoRegression

feature_dim = auto_regression_dataset[0][1][1][0].shape[-1]
model = AutoRegression(feature_dim=feature_dim, hidden_dim=512, recursive_layers=8, dtype=DTYPE)
model, model.dtype, model.device
```

# 模型训练

```python
import torch
from torch import nn, optim

num_epochs = 100_000_000
elastic_net_param = {'alpha': 1e-4, 'rho': 0.2}
learning_rate = 2e-6
train_loss_threshold = 0.
valid_loss_threshold = 0.
criterion = nn.CrossEntropyLoss()
optim = optim.Adamax(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0
    for batch_inputs, batch_labels in auto_regression_dataset_train_loader:
        outputs = torch.softmax(model.forward(batch_inputs, model.initial_state(batch_size=batch_inputs.shape[0]))[0], dim=-1)
        loss = criterion(outputs, batch_labels) + model.elastic_net(alpha=elastic_net_param['alpha'], rho=elastic_net_param['rho'])
        optim.zero_grad()
        loss.backward()
        optim.step()
        total_loss += loss.item()
    if epoch % 3 == 0:
        total_valid_loss = 0.0
        for batch_inputs, batch_labels in auto_regression_dataset_valid_loader:
            with torch.no_grad():
                model.eval()
                outputs = torch.softmax(model.forward(batch_inputs, model.initial_state(batch_size=batch_inputs.shape[0]))[0], dim=-1)
                loss = criterion(outputs, batch_labels) + model.elastic_net(alpha=elastic_net_param['alpha'], rho=elastic_net_param['rho'])
                total_valid_loss += loss.item()
                model.train()
        print(f"Epoch {epoch + 1}/{num_epochs} | "
              f"Train Loss: {total_loss:.4f} | "
              f"Valid Loss: {total_valid_loss:.4f}")
        if total_valid_loss <= valid_loss_threshold:
            break
    print(f"Epoch {epoch + 1}/{num_epochs} | Train Loss: {total_loss:.4f}")
    if total_loss <= train_loss_threshold:
        break
```

# 模型测试

```python
from data.note_digitize import decode as d_decode
from data.note_vectorize import decode as v_decode

model.eval()
for batch_inputs, batch_labels in auto_regression_dataset_valid_loader:
    with torch.no_grad():
        print('Input: ')
        for b in batch_inputs:
            print([d_decode(v_decode(x)) for x in b])
        outputs = torch.softmax(model.forward(batch_inputs, model.initial_state(batch_size=batch_inputs.shape[0]))[0], dim=-1)
        outputs = torch.argmax(outputs, dim=-1)
        print('Predict: ')
        for b in batch_inputs:
            print([d_decode(x) for x in outputs])
        print('-' * 36)
```
